# Probablistic Optimization
This notebook approaches similar problems as the previous notebook, using probability to estimate the profit for each given price. This is quick, computed optimization & hopes to lead to the closed form solution

```python
'''
This helper function inputs an unsorted array of prices, costs, tiers, and the scaling_param (lambda) and will 
output the intervals for which each price/cost combination is the maximum
'''
def calculate_intervals(prices_unsorted, costs_unsorted, tiers, scaling_param):
    
    # We need to argsort by cost first because function logic relies on sequence of increasing slope lines (slope is cost)
    sorted_indices = np.argsort(costs_unsorted)
    costs = np.array(costs_unsorted)[sorted_indices]
    prices = np.array(prices_unsorted)[sorted_indices]

    thresholds = []
    thresholds.append(-sys.float_info.max)
    
    for i in range(tiers):
            if i == 0:
                thresholds.append(prices[0]/costs[0])
                                  
            else:
                intersection = (prices[i]-prices[i-1])/(costs[0] * ((costs[i]/costs[0])**(scaling_param)-(costs[i-1]/costs[0])**(scaling_param)))
                j = 0
                while intersection < thresholds[i - j]:
                    j += 1
                    if i == j:
                        intersection = prices[i]/(costs[0]*(costs[i]/costs[0])**(scaling_param))
                        break;
                    else:
                        intersection = (prices[i]-prices[i-j-1])/(costs[0] * ((costs[i]/costs[0])**(scaling_param)-(costs[i-j-1]/costs[0])**(scaling_param)))
                    
                thresholds.append(intersection)
                
                
    for i in range(tiers):
        if (thresholds[tiers - i] < thresholds[tiers-i-1]):
            thresholds[tiers-i-1] = thresholds[tiers - i]
            
    intervals = []
    for i in range(tiers):
        intervals.append((thresholds[i],thresholds[i+1]))
    
    intervals.append((thresholds[len(thresholds)-1],sys.float_info.max))

    intervals_ordered = [None] * len(intervals)
    intervals_ordered[0] = intervals[0]
    for i, idx in enumerate(sorted_indices):
        intervals_ordered[idx+1] = intervals[i+1]

    return intervals_ordered
```

```python
class tiered_pricing_system:
    def __init__(self, costs, tiers, scaling_param, mu, sigma, pdf_type = 'uniform'):
        self.costs = costs
        self.tiers = tiers
        self.scaling_param = scaling_param
        self.mu = mu
        self.sigma = sigma
        self.pdf_type = pdf_type
        
    def tier_probabilities(self, prices):
        probabilities = []
        intervals = calculate_intervals(prices, self.costs, self.tiers, self.scaling_param)

        if (self.pdf_type == 'uniform'):
            start = self.mu - self.sigma
            end = self.mu + self.sigma
            point_prob = 1/(end - start)
            for i in range(self.tiers+1):
                probabilities.append(max((min(intervals[i][1],end)-max(intervals[i][0],start))*point_prob,0))
        elif (self.pdf_type == 'gaussian'):
            for i in range(self.tiers+1):
                probabilities.append(norm.cdf(intervals[i][1],self.mu,self.sigma) - norm.cdf(intervals[i][0],self.mu,self.sigma))
        else:
            raise ValueError('pdf_type must be uniform or gaussian')

        return probabilities
        
        
    def profit(self, prices):
        probabilities = self.tier_probabilities(prices)

        profits = [pr * (p - c) for (p, c, pr) in zip(prices, self.costs, probabilities[1:])]
        return sum(profits)
```

```python
# negative of profit because scipy likes to minimize
def objective(prices, system):
    return -system.profit(prices)
```

# Over Time Based Approach

This first trial is a failure. It uses coordinate descent, which fails to take into account the customers' decision makings. Prices need to be increased in conjunction with each other, which likely means gradient descent is necessary

## Gradient Descent
Inspired by "Network-aware Recommender System via Online Feedback Optimization", I'll implement a gradient descent where we are aware of all of the parameter states to try to optimize. I'll use this as a baseline to try to figure out how to estimate gradients in later steps.

```python
class gradient_descent:
    
    def __init__(self, system, tolerance = 1e-6, max_iters = 1000, gradient_delta = 1e-6, patience=2):
        self.system = system
        #self.learning_rate = min(system.costs)
        self.learning_rate = min(system.costs) / 5
        self.tolerance = tolerance
        self.max_iters = max_iters
        self.gradient_delta = gradient_delta
        self.patience=patience
        

    # Function to approximate gradient
    def numerical_gradient(self, prices):
        grad = [0] * len(prices)
        for i in range(len(prices)):
            vec = [0] * len(prices)
            vec[i] = 1

            grad[i] = (system.profit(prices + self.gradient_delta * np.asarray(vec)) 
                       - system.profit(prices)) / self.gradient_delta
        return grad
    
    def maximize(self):
        self.prices = system.costs.copy()
        self.profit = system.profit(self.prices)
        self.profit_history = [self.profit]
        self.price_history = [self.prices]
        patience_count=0
        for i in range(self.max_iters):
            grad = self.numerical_gradient(self.prices)
            prices_next = self.prices + np.asarray(self.learning_rate) * grad
            
            # If gradient is close enough to zero, we're done ; early stopping
            if np.linalg.norm(prices_next - self.prices) < self.tolerance:
                break;
                
            self.prices = prices_next
            self.profit = system.profit(prices_next)
            self.profit_history.append(self.profit)
            
            self.price_history.append(prices_next.copy())
            
            # reduce the learning rate if we are going in the wrong direction
            if (self.profit_history[len(self.profit_history) - 1] < 
                self.profit_history[len(self.profit_history) - 2]):
                patience_count += 1
                if(patience_count == self.patience):
                    self.learning_rate *= 0.5
                    patience_count = 0
```
